```python
response_cid = requests.get("https://raw.githubusercontent.com/davidgasquez/gitcoin-grants-data-portal/main/data/IPFS_CID")
CID = response_cid.text.replace("\n", "")
CID
```

```python
file = "round_applications"
url = "https://ipfs.filebase.io/ipfs/{}/{}.parquet".format(CID, file)
url
```

```python
connection = duckdb.connect()
cursor = connection.cursor()
sql_query = """
SELECT * FROM '{}' LIMIT 20;
""".format(url)

cursor.execute(sql_query)

results = cursor.fetchall()
```
